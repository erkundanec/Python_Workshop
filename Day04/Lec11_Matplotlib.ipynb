```python
import numpy as np
import matplotlib.pyplot as plt
```

# Line Plot

```python
t=0.2
xx=np.linspace(0,t,1000)
freq=50
yy=np.sin(2*np.pi*freq*xx)
```

```python
plt.plot(xx,yy)
plt.xlabel('time')
plt.ylabel('current')
plt.title('current vs time')
plt.savefig('my_first_plot.png',dpi=300)
```

## Plotting data from text file

```python
x,y=np.loadtxt('data.txt',delimiter='\t',unpack=True)
print(len(x))
```

```python
plt.plot(x[0::5],y[0::5],'k<')
plt.plot(x,x**2,'r-')
```

Practice: Place xlabel and ylabel on the figure

```python
plt.plot(x[0::5],y[0::5],'k^',label='data')
plt.plot(x,x**2,'r-',label='function')
plt.legend(edgecolor='none')
```

## Subplot

```python
plt.subplot(211)
plt.plot(xx,yy)
plt.xlabel('time')
plt.ylabel('current')

plt.subplot(212)
plt.plot(x,y)
plt.xlabel('x-values')
plt.ylabel('y-values')
```

```python
plt.subplot(211)
plt.plot(xx,yy)
plt.xlabel('time')
plt.ylabel('current')

plt.subplot(212)
plt.plot(x,y)
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.tight_layout(pad=2.0)
plt.savefig('subplot.pdf',dpi=300)
```

# Advance Concepts (3D plot)

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.pyplot import figure,show
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
fig=plt.figure()
ax = fig.gca(projection='3d')

xrange=np.linspace(-5,5,100)
yrange=np.linspace(-5,5,100)

x3d,y3d=np.meshgrid(xrange,yrange)
f=0.2

r=np.sqrt(x3d**2+y3d**2)
z3d=np.sin(2*np.pi*f*r)
surf = ax.plot_surface(x3d, y3d, z3d, cmap=cm.plasma)

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.show()
```

Available colormaps are viridis, magma, inferno, plasma, etc.
